```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
doge_trend = TrendReq()
```

After importing all package, here I will be analyzing the Google search trends on the queries based on “Dogecoin”, so let’s create a DataFrame of the top 10 countries which search for “Dogecoin” on Google Search

```python
doge_trend.build_payload(kw_list=["Dogecoin"])
data_doge = doge_trend.interest_by_region()
data_doge = data_doge.sort_values(by="Dogecoin", ascending=False)
data_doge = data_doge.head(10)
print(data_doge)
```

We can see now this are top 10 coutries inside google search that are always query about "Dogecoin" . Surprisingly we can see that Singapore also listed as a top 10 countries that always talk about "Dogecoin"

```python
data_doge.reset_index().plot(x="geoName", y="Dogecoin", 
                        figsize=(20,18), kind="bar")
plt.style.use('fivethirtyeight')
plt.show()
```

We visualize it in a bar chart in order for people to have a good looks and to have better understanding in what i'm trying to story right now.

```python
data_doge = TrendReq(hl='en-US', tz=360)
data_doge.build_payload(kw_list=['Dogecoin'])
data_doge = data_doge.interest_over_time()
fig, ax = plt.subplots(figsize=(20, 15))
data_doge['Dogecoin'].plot()
plt.style.use('fivethirtyeight')
plt.title('Total Google Searches for Dogecoin', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

Wah surprisingly from the line chart, we can see that nobody care about "Dogecoin" before 2021.

Start from 2021, we can see that a lot of people start to find information regarding "Dogecoin" and it is very surprise.
